## Lec 10 - Indexing and Selectiong data

```python
# Selecting Data

# Select Columns
df["country"]
df[["city", "aqi"]]
```

```python
# Select rows (by label)

df.loc[0]            # 1st row (by lable)
df[0:3]              # row0 to row3 - both inclusive in loc
```

```python
# Select row (by indexing)
df.iloc[0]           # 1st row
df.iloc[4:7]         # 2st row
```

```python
# Select row & columns

df.loc[0, "country"]    # 0th row & specific column
df.iloc[0, 2]           # 1st row & 3rd column (by position / indexing)
```

```python
df.iloc[0, 1:5]
df.loc[0, ["country", "aqi"]]
```

```python
df.loc[0:5, ["country","city","latitude"]]  # end is inclusive
df.iloc[0:5, 1:3]                           # end is exclusive
```

```python
# Select single scalar value
df.at[0, "city"]
df.iat[0, 2]
```

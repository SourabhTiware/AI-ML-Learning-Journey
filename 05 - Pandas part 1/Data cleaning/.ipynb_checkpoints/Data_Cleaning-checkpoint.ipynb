## Lec 13 - Data Cleaning - Handle Missing Values

```python
df = pd.read_csv("raw_data.csv")
```

```python
df.isnull() # Where values is None set as True. 
df.isna()   # it is alias name of the isnull() method. 
```

```python
df
```

```python
df.isnull().sum()  # it is count total values are None in single column. 
# if total 3 values are None in age column it store 3 age column. 
```

```python
df.dropna() # It is a drop row is null value in  row. 
```

```python
# df.dropna(axis=1) # drop a column those have None value in column.
```

```python
df.fillna(0) # it is fill values where None value avaiable. 
```

```python
df.ffill() # Fill forward values 
# means => in age column "Jane smit" has no age value in age column. "ffill() => method " fill the farward value of "John Doe" age. 
```

```python
df.bfill() # vise varsa like ffill=> forward fill. It start from the backword values. 
```

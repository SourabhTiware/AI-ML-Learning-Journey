## Lec 01 - Data Transformation

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("raw_data.csv")
df2 = df.copy()
```

```python
# Data Transformation 
# .apply() method add new column in existing data.

df2["tax"] = df2["income"].apply(lambda x: "20%" if x >= 60000 else "10%")
df2
```

```python
df2 = df2.ffill()
df2
```

```python
# map method - 
# .map() method ietrate over an column values and update those values. 

gender_map = {"Male": "M", "Female" : "F", "Unknown" : "U"}
df2["gender"] = df2["gender"].map(gender_map)
```

```python
df2
```

```python
# .assign() - using assign method we create a new column in existing data. 

df2.assign(new_income = df2["income"] * 1.1)
```

```python
# .replace(old_vale, new_value) => replace value in column. 

df2["country"] = df2["country"].replace("USA", "United States of America")
```

```python
df2
```

## Lec 02 -Data Transformation - Other Methods

```python
# Renameing

df2.columns = ["Id", "Name", "Age", "Country", "Gender", "Income", "Tax"]
df2.rename(columns={"Income":"Salary"})
df2.rename(index={0:"first"})
```

```python
# Sorting - values & index
df2.sort_values("Income") # # sort values in ascending
df2.sort_values("Income", ascending=False) # sort values in Desending 
df2.sort_values(["Age","Income"])       # sorts for age, if age is same then sorts for income 
```

```python
sorted_df2 = df2.sort_values(["Income", "Age"])
sorted_df2.sort_index()
```

```python
# Reset Index
sorted_df2.reset_index() # Pandas here maintain the previous index and add extra column index and maintain the previous index
# but we want to new index value and remove pandas maintaiale index then use following command 
sorted_df2.reset_index(drop = True)
```

```python
# Ranking the data according to Income, Age, points and anything
df2["Ranking"] = df2["Income"].rank(ascending=False, method="dense")

# df2["Ranking"] = df2["Income"].rank(ascending=False, method="max")
# df2["Ranking"] = df2["Income"].rank(ascending=False, method="min")


df2
```

```python
# Re-order 
df2[["Name", "Age",	"Country" ,"Gender", "Income", "Ranking", "Tax"]]
```

## Lec - 05 Grouping & Aggregation

```python
new_df = pd.read_csv("sorted_data.csv")
new_df = new_df.drop(columns=["Unnamed: 0"])

df2 = new_df.copy()
df2
```

```python
df2.groupby("country")["income"].mean() # mean income for each country
df2.groupby("gender")["income"].mean() # mean income for each gender
```

```python
df2.groupby("country")["gender"].count() # count of gender for each country
df2.groupby("gender")["income"].max()    # max income for each gender
df2.groupby("gender")["income"].min()    # min income for each gender
```

```python
df2.groupby("country")["income"].agg(["mean","max","min"])  # applies multiple aggregate functions
df2.groupby("country")["income"].aggregate(["mean","max","min"]) # alies of agg method
```

```python
 # rename aggregate
df2.groupby("country")["income"].agg(avg_salary="mean", max_salary="max", min_salary="min") 
```

```python
# aggregate multiple cols
df2.groupby("country").agg({
    "age":"mean",
    "income":"mean"
})
```

```python
df2.groupby("country").agg(
    avg_age=("age","mean"),
    avg_income=("income","mean")
)
```

## Lec-06 Melt & Pivot for Reshaping

```python
df_melt_pivot = pd.DataFrame({
    "country":["USA", "USA", "India", "India"],
    "year":[2020, 2021, 2020, 2021],
    "sales":[100, 120, 90, 110],
    "profit":[20,25,18,22]
})
df_melt_pivot
```

```python
melted_df = df_melt_pivot.melt(
    id_vars=["country","year"],
    value_vars=["sales","profit"],
    var_name="metric",
    value_name="value (in Lakhs)"
)
melted_df
```

```python
original = melted_df.pivot(
    index=["country","year"],
    columns="metric",
    values="value (in Lakhs)"
)
original
```

## Load a data from stroe_data.json file

```python
import json
```

```python
def load_data(filename):
    with open(filename, "r") as f:
        data = json.load(f)

    return data
```

```python
data = load_data("store_data.json")
print(data )
print(type(data))

```

```python
# Clearn and Structring data

def clean_data(data):
    
    text_to_num = {"one":1, "two":2, "three": 3, "four": 4, "five": 5}

    clean_data = []
    unique_users = set()
    
    for user in data:
        #Clean Rating
        raw_rating = str(user["rating"]).strip().lower()
        if(raw_rating in text_to_num):
            raw_rating = text_to_num[raw_rating] 
        user["rating"]= raw_rating
        # print(user)

        # Handle Missing values

        raw_age = user.get("age")
        if(raw_age == None):
            user["age"] = None

        # Deduplication
        if(user["name"].strip() in unique_users):
            continue
    
        unique_users.add(user["name"])
        clean_data.append(user)
   
    return clean_data

 
        
```

```python
data = clean_data(data)
```

```python
# Calculate Average Rating

def meaning_insight(data):
    total_rat = 0
    for user in data:
        total_rat += float(user["rating"])
    print(f"total average rating is {total_rat / len(data)}")

    # % of user give poor rating
    poor_rat = 0
    for user in data:
        if(float(user["rating"] ) < 3):
            poor_rat += 1

    print(f"total % of poor rating {poor_rat / len(data) * 100}%")
```

```python
meaning_insight(data)
```

```python
# Recommendation feature

def get_recommendation(data):
    recommendation = []
    # print(data)
    for user in data:
        curr_recomm = {}
        curr_recomm['name'] = user['name']
        if(float(user['rating']) >= 4):
            curr_recomm['brand'] = 'Apple'
        else:
            curr_recomm['brand'] = 'samsung'
            
        recommendation.append(curr_recomm)
    return recommendation
```
